```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
xs = np.arange(0.000001, 1.1, 0.001)
```

```python
xs2 = np.ones(xs.size)-xs
```

```python
f = lambda x: -np.log(x)
ys1 = f(xs)
ys2 = np.log(xs)
ys3 = f(xs2)
```

```python
fig, ax = plt.subplots()
ax.plot(xs, ys1, 'blue', label='-log', alpha=0.5)
ax.plot(xs, ys2, 'red', label='+log', alpha=0.5)
ax.scatter(1, 0, color='black', marker='x', label='(1,0)', alpha=1)
ax.legend()
ax.grid(True)
plt.show
```

```python
fig, ax = plt.subplots()
ax.plot(xs, ys1, 'blue', label='-log(x)', alpha=0.5)
ax.plot(xs, ys3, 'red', label='-log(1-x)', alpha=0.5)
ax.scatter(1, 0, color='black', marker='x', label='(1,0)', alpha=1)
ax.legend()
ax.grid(True)
plt.show
```

```python
xs = np.arange(-5, 5, 0.1)
g = lambda x: 1/(1 + np.exp(-x))
ys = g(xs)
```

```python
fig, ax = plt.subplots()
ax.plot(xs, ys, 'blue', label='sigmoid', alpha=0.5)
ax.scatter(0, 0.5, color='black', marker='x', label='(0,0.5)', alpha=1)
ax.legend()
ax.grid(True)
plt.savefig('sigmoid.png')
```
